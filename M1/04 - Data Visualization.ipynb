Ejercicio 1. Carga el conjunto de datos "iris" de Seaborn en un DataFrame de Pandas. Luego, crea un gráfico de dispersión (scatter plot) utilizando Matplotlib que muestre la relación entre la longitud del sépalo ("sepal_length") y la longitud del pétalo ("petal_length"). Etiqueta los ejes y usa diferentes colores para cada especie de iris.

````python
# Cargar los datos del dataset iris
iris = sns.load_dataset("iris")

# Definir los datos para el gráfico
sepal_length = iris["sepal_length"]
petal_length = iris["petal_length"]
species = iris["species"]

# Crear el scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=sepal_length, y=petal_length, hue=species, palette="Set1", s=100)

# Agregar etiquetas a los ejes y título
plt.xlabel("Longitud del Sépalo (cm)")
plt.ylabel("Longitud del Pétalo (cm)")
plt.title("Relación entre Longitud del Sépalo y Longitud del Pétalo")

# Mostrar el gráfico
plt.legend(title="Especie")
plt.grid(True)
plt.show()
````

Ejercicio 2. Utiliza Pandas para cargar el conjunto de datos "tips" de Seaborn en un DataFrame. Luego, calcula el promedio de la factura total ("total_bill") para cada día de la semana y crea un gráfico de barras (bar plot) que muestre estos promedios utilizando

````python
# Cargar el conjunto de datos tips
tips = sns.load_dataset("tips")

# Calcular el promedio de la factura total ("total_bill") para cada día de la semana
promedio_por_dia = tips.groupby('day')['total_bill'].mean().reset_index()

# Crear el bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='total_bill', data=promedio_por_dia, palette='muted')

# Agregar etiquetas a los ejes y título
plt.xlabel('Día de la semana')
plt.ylabel('Promedio de la factura total ($)')
plt.title('Promedio de la factura total por día de la semana')

# Mostrar el gráfico
plt.show()
````

Ejercicio 6. Dado un Dataframe de salarios, obteng el segundo salario mas alto del DataFrame Employee. Si no hay segundo mas alto, que la funcion retorne valro nulo.



```
Employee table:
+----+--------+
| id | salary |
+----+--------+
| 1  | 100    |
| 2  | 200    |
| 3  | 300    |
+----+--------+
```



````python
#Dejo declarado como es el Dataframe
df_Employee = pd.DataFrame({'id': [1, 2, 3,4,5,6,7,8,9],
        'salary': [100, 200, 300,800,900,1230,120,120,1230]})
````

````python
def segundo_salario_mas_alto(df):
    # Encontrar el salario más alto
    salario_mas_alto = df['salary'].max()

    # Eliminar el salario más alto y encontrar el siguiente más alto
    segundo_salario = df['salary'].nlargest(2).iloc[-1]

    # Si el segundo salario es igual al máximo, no hay segundo más alto
    if segundo_salario == salario_mas_alto:
        return None
    else:
        return segundo_salario

# Obtener el segundo salario más alto
segundo_salario = segundo_salario_mas_alto(df_Employee)
print("El segundo salario más alto es:", segundo_salario)
````

Ejercicio 7. Escriba una solución que encuentre los empleados que ganan mas que sus Managers.

Input



```
+----+-------+--------+-----------+
| id | name  | salary | managerId |
+----+-------+--------+-----------+
| 1  | Joe   | 70000  | 3         |
| 2  | Henry | 80000  | 4         |
| 3  | Sam   | 60000  | Null      |
| 4  | Max   | 90000  | Null      |
+----+-------+--------+-----------+
```

Output

```
+----------+
| Employee |
+----------+
| Joe      |
+----------+
```




````python
data = {'id': [1, 2, 3, 4],
        'name': ['Joe', 'Henry', 'Sam', 'Max'],
        'salary': [70000, 80000, 60000, 90000],
        'managerId': [3, 4, None, None]}

df = pd.DataFrame(data)
````

````python
# Agregar una columna 'manager_salary' que contenga los salarios de los gerentes
df_merged = df.merge(df[['id', 'salary']], left_on='managerId', right_on='id', how='left', suffixes=('', '_manager'))

# Filtrar los empleados que ganan más que sus gerentes
empleados_que_ganan_mas_que_manager = df_merged[df_merged['salary'] > df_merged['salary_manager']]

# Renombrar la columna 'name' como 'Employee'
empleados_que_ganan_mas_que_manager = empleados_que_ganan_mas_que_manager.rename(columns={'name': 'Employee'})

# Mostrar solo la columna 'Employee'
output = empleados_que_ganan_mas_que_manager[['Employee']].reset_index(drop=True)

print(output)
````
