Ejercicio 2. Escribe una función llamada cuadrados que tome una lista de números como argumento y devuelva una lista de los cuadrados de esos números.

````python
def cuadrados(list):
    lista_cuadrados = []
    for item in list:
        lista_cuadrados.append(item**2)
    return lista_cuadrados
````

Ejercicio 3. Escribe una función llamada es_palindromo que tome una cadena como argumento y devuelva True si la cadena es un palíndromo (se lee igual de izquierda a derecha y de derecha a izquierda), y False en caso contrario.

````python
def es_palindromo(string):
  lista = []
  for i in string:
    lista.insert(0,i)
  cadena = ''.join(lista)
  if cadena == string:
    return True
  else:
    return False

````

Ejercicio 12. Escribe una solución para calcular el tiempo total en minutos que cada empleado pasó en la oficina cada día. Ten en cuenta que en un solo día, un empleado puede entrar y salir más de una vez. El tiempo que un empleado pasó en la oficina para una sola entrada es igual a la hora de salida (out_time) menos la hora de entrada (in_time). Si el resultado son tiempos negativos cambiar ese valor por un 0 (cero). Devuelve la tabla de resultados en cualquier orden.

El formato del resultado es como se muestra en el siguiente ejemplo.





```
+------------+--------+------------+
| day        | emp_id | total_time |
+------------+--------+------------+
| 2020-11-28 | 1      | 173        |
| 2020-11-28 | 2      | 30         |
| 2020-12-03 | 1      | 41         |
| 2020-12-09 | 2      | 27         |
+------------+--------+------------+
```




````python
import random
from datetime import datetime, timedelta

# Crear el DataFrame original
data = {
    'emp_id': [1, 1, 1, 2, 2],
    'event_day': ['2020-11-28', '2020-11-28', '2020-12-03', '2020-11-28', '2020-12-09'],
    'in_time': [4, 55, 1, 3, 47],
    'out_time': [32, 200, 42, 33, 74]
}

df = pd.DataFrame(data)

# Generar 40 filas adicionales con datos aleatorios
additional_data = []

for _ in range(40):
    emp_id = random.randint(1, 5)
    event_day = (datetime(2020, 11, 28) + timedelta(days=random.randint(0, 15))).strftime('%Y-%m-%d')
    in_time = random.randint(0, 240)
    out_time = in_time + random.randint(1, 240)
    additional_data.append([emp_id, event_day, in_time, out_time])

additional_df = pd.DataFrame(additional_data, columns=['emp_id', 'event_day', 'in_time', 'out_time'])

# Concatenar el DataFrame original y el DataFrame adicional
df = pd.concat([df, additional_df], ignore_index=True)

# Mostrar el DataFrame resultante
print(df.head())


## Aca debajo se puede comenzar a resolver el ejercicio.
````

````python
df['total_time'] = df['out_time'] - df['in_time']
df['total_time'] = df['total_time'].mask(df['total_time'] < 0, 0)
columnas = ['event_day','emp_id','total_time']
df_salida = df[columnas]
df_salida = df_salida.rename(columns={'event_day': 'day'})
print(df_salida.head())
````
