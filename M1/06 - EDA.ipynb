```python
#Podemos usar un poco de cÃ³digo para visualizar mÃ¡s de una variable a la vez
variables = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
            'pH', 'sulphates', 'alcohol', 'quality']

columns = 4

fig, axes = plt.subplots(len(variables) //columns, columns, figsize=(15,8))

for current_idx, variable in enumerate(variables):
    i = current_idx // columns
    j = current_idx % columns
    sns.distplot(wine_quality[variable], ax=axes[i][j])
    axes[i][j].set_title(variable)
    axes[i][j].set_xlabel("")

plt.tight_layout()

#Density = Porcentaje de vinos (%)
```

De esta grÃ¡fica podemos identificar varios aspectos, por ejemplo, que muchas de las variables tienen una distribuciÃ³n asimÃ©trica (fixed acidity, residual sugar, chlorides, por ejemplo), ademÃ¡s de que al parecer algunos valores presetan valores extremos (residual sugar, sulphates, total sulfur dioxide).

# Boxplots
Como sabemos los boxplots nos permiten analizar y ver de una manera muy rÃ¡pida quÃ© variables tienen valores extremos.
Recordamos que: "cualquier dato que estÃ© mÃ¡s allÃ¡ de 3 veces el valor del desvÃ­o estÃ¡ndar alejado del promedio puede considerarse un valor extremo o atÃ­pico"

```python
variables = ['fixed acidity', 'residual sugar',
             'chlorides', 'free sulfur dioxide',
             'total sulfur dioxide','sulphates', 'alcohol']

fig, axes = plt.subplots(1, len(variables), figsize=(15,6))

for ax, variable in zip(axes, variables):
    ax = sns.boxplot( y=variable, data=wine_quality, ax=ax)
plt.tight_layout()
```

# Scatterplots
Ahora tratemos de identificar de una manera rÃ¡pida y visual la correlaciÃ³n entre variables!

```python
from seaborn import lmplot
lmplot(x='quality', y='alcohol', data=wine_quality, fit_reg=True)
```

En la grÃ¡fica anterior, si se puede observar una clara relaciÃ³n entre las variables analizadas. TambiÃ©n podemos
hacer uso de la Matriz de CorrelaciÃ³n si quisieramos ver por ejemplo el coeficiente de **Pearson**

```python
correlation = wine_quality.corr(method="pearson")
correlation
```

La matriz de correlaciÃ³n muestra los valores de correlaciÃ³n de Pearson, que miden el grado de relaciÃ³n lineal entre cada par de elementos o variables. Los valores de correlaciÃ³n se pueden ubicar entre -1 y +1.
Para interpretar los valores del coeficiente de correlaciÃ³n de Pearson, se pueden utilizar los siguientes criterios:
Entre 0 y 0,10: correlaciÃ³n inexistente
Entre 0,10 y 0,29: correlaciÃ³n dÃ©bil
Entre 0,30 y 0,50: correlaciÃ³n moderada
Entre 0,50 y 1,00: correlaciÃ³n fuerte

```python
#Resulta mÃ¡s intuitivo ver la matriz de correlaciÃ³n a travÃ©s de un grÃ¡fico - heatmap de seaborn

plt.figure(figsize=(10,10)) #Defino el tamaÃ±o del grafico

#Genero el mapa de calor
sns.heatmap(correlation,
            cbar = True,
            square = True,
            annot=True,
            fmt= '.2f',
            annot_kws={'size': 15},
            cmap= 'coolwarm')

#Roto los ejes para que queden inclinados
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.show()
```
